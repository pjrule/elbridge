```python
import os
import pytest
import numpy as np
import geopandas as gpd
from math import pi, sqrt
from elbridge import Bitmap
from elbridge.bitmap import _mask
from shapely.geometry import box
from geopandas import GeoDataFrame

FIXTURES_DIR = os.path.join(os.path.dirname('.'), 'fixtures')
FIXTURE_FILES = {
    'pa129': os.path.join('pa129', 'pa129.shp')
}

def pa129():
    pa = gpd.read_file(os.path.join(FIXTURES_DIR, FIXTURE_FILES['pa129']))
    pa = pa.to_crs({'init': 'esri:102003'})
    meta = {
        'width': pa.bounds['maxx'].max() - pa.bounds['minx'].min(),
        'height': pa.bounds['maxy'].max() - pa.bounds['miny'].min()
    }
    meta['alpha'] = meta['width'] / meta['height']
    # TODO pick reasonable resolutions
    return Bitmap(pa, 'TOT_POP', 400, 1), meta
```

```python
pa = pa129()[0]
```
