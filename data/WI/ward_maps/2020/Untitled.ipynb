```python
import geopandas as gpd
gdf = gpd.read_file('2012_2020_Election_Data_with_2017_Wards.shp').to_crs({'init': 'esri:102003'})
```

```python
import numpy as np
pops = np.array(gdf['PERSONS'])
areas = np.zeros(len(gdf))
for idx, row in gdf.iterrows():
    areas[idx] = row.geometry.area
pop[pop==0] = 1
log_density = np.log(pop / areas)
gdf['log_pop'] = log_density
gdf['density'] = pop / areas
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors

fig, ax = plt.subplots(figsize=(8,8))
gdf.plot(ax=ax, column='density', legend=True,
         norm=colors.LogNorm(vmin=gdf['density'].min(), vmax=gdf['density'].max()))

plt.axis('on')
plt.savefig('pop_density.png', dpi=300, transparent=True, bbox_inches='tight')
plt.show()
```
